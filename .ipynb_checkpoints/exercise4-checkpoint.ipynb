```python
from fastai.vision import *
from fastai.callbacks.hooks import *
from fastai.utils.mem import *
```

```python
path = untar_data(URLs.CAMVID)
path.ls()
```

```python
path_lbl = path/'labels'
path_img = path/'images'
```

```python
fnames = get_image_files(path_img)
fnames[:3]
```

```python
lbl_names = get_image_files(path_lbl)
lbl_names[:3]
```

```python
img_f = fnames[2]
img = open_image(img_f)
img.show(figsize=(5,5))
```

```python
get_y_fn = lambda x: path_lbl/f'{x.stem}_P{x.suffix}'
```

```python
mask = open_mask(get_y_fn(img_f))
mask.show(figsize=(5,5), alpha=1)
```

```python
src_size = np.array(mask.shape[1:])
src_size,mask.data
```

```python
codes = np.loadtxt(path/'codes.txt', dtype=str); codes
```

## 2) camvid 데이터셋

```python
size = src_size//2

free = gpu_mem_get_free_no_cache()
# the max size of bs depends on the available GPU RAM
if free > 8200: bs=8
else:           bs=4
print(f"using bs={bs}, have {free}MB of GPU RAM free")
```

```python
src = (SegmentationItemList.from_folder(path_img)
       .split_by_fname_file('../valid.txt')
       .label_from_func(get_y_fn, classes=codes))
```

```python
data = (src.transform(get_transforms(), size=size, tfm_y=True)
        .databunch(bs=bs)
        .normalize(imagenet_stats))
```

```python
data.show_batch(5, figsize=(10,7))
```

```python
data.show_batch(2, figsize=(10,7), ds_type=DatasetType.Valid)
```

## 3) camvid 데이터 모델

```python
name2id = {v:k for k,v in enumerate(codes)}
void_code = name2id['Void']

def acc_camvid(input, target):
    target = target.squeeze(1)
    mask = target != void_code
    return (input.argmax(dim=1)[mask]==target[mask]).float().mean()
```

```python
metrics=acc_camvid
```

```python
wd=1e-2
```

```python
learn = unet_learner(data, models.resnet34, metrics=metrics, wd=wd)
```

```python
lr_find(learn)
learn.recorder.plot()
```

```python
lr=3e-3
```

```python
learn.fit_one_cycle(1, slice(lr), pct_start=0.9)
```

```python
learn.save('stage-1')
```

```python
learn.load('stage-1');
```

```python
learn.show_results(rows=3, figsize=(8,9))
```

```python

learn.unfreeze()
```

```python
lrs = slice(lr/400,lr/4)
```

```python
learn.fit_one_cycle(12, lrs, pct_start=0.8)
```

```python
learn.save('stage-2');
```

```python
learn.destroy()

size = src_size

free = gpu_mem_get_free_no_cache()
# the max size of bs depends on the available GPU RAM
if free > 8200: bs=3
else:           bs=1
print(f"using bs={bs}, have {free}MB of GPU RAM free")
```

```python
data = (src.transform(get_transforms(), size=size, tfm_y=True)
        .databunch(bs=bs)
        .normalize(imagenet_stats))
```

```python
learn = unet_learner(data, models.resnet34, metrics=metrics, wd=wd)
```

```python
learn.load('stage-2');
```

```python
lr_find(learn)
learn.recorder.plot()
```

```python
lr=1e-3
```

```python
learn.fit_one_cycle(4, slice(lr), pct_start=0.8)
```

```python
learn.save('stage-1-big')
```

```python
learn.load('stage-1-big');
learn.unfreeze()
lrs = slice(1e-6,lr/10)
```

```python
learn.fit_one_cycle(5, lrs)
```

```python
learn.save('stage-2-big')
```

```python
learn.load('stage-2-big');
```

```python
learn.show_results(rows=3, figsize=(10,10))
```

# 4. Regression with BIWI head pose dataset

> ## 이 예제에서는 저번 시간에 배운 regression이 어떻게 적용되는 지를 한 번 생각해보세요.
> ## regression과 classification의 차이는 무엇이며
> ## loss function은 그럼 무엇을 의미하는지 등을요.
> ## 그리고 그것들과 get_ctr의 관계를 한 번 생각해보세요

* 이미지 회귀방식으로 머리의 중심을 찾는 예제이다.

## 1) 데이터 얻고 바꾸기

```python



path = untar_data(URLs.BIWI_HEAD_POSE)
```

```python
cal = np.genfromtxt(path/'01'/'rgb.cal', skip_footer=6); cal
```

```python




fname = '09/frame_00667_rgb.jpg'
```

```python
def img2txt_name(f): return path/f'{str(f)[:-7]}pose.txt'
```

```python



img = open_image(path/fname)
img.show()
```

```python
ctr = np.genfromtxt(img2txt_name(fname), skip_header=3); ctr
```

```python
def convert_biwi(coords):
    c1 = coords[0] * cal[0][0]/coords[2] + cal[0][2]
    c2 = coords[1] * cal[1][1]/coords[2] + cal[1][2]
    return tensor([c2,c1])

def get_ctr(f):
    ctr = np.genfromtxt(img2txt_name(f), skip_header=3)
    return convert_biwi(ctr)

def get_ip(img,pts): return ImagePoints(FlowField(img.size, pts), scale=True)
```

```python
get_ctr(fname)
```

```python
ctr = get_ctr(fname)
img.show(y=get_ip(img, ctr), figsize=(6, 6))
```
