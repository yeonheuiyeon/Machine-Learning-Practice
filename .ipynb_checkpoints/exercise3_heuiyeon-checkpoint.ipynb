## 8. Linear Regression problem

신경망에서 대체적으로 사용되는 SGD에대해서 배워보기

- 선형 회귀 분석의 목표는 점 집합에 선을 맞추는 것이다.

````python
n=100
````

````python
x = torch.ones(n,2) 
x[:,0].uniform_(-1.,1)
x[:5]
````

- tensor 는 어레이 종류라고 하셨다. 따라서 []의 갯수에 따라 몇차원 배열인지 알수 있다. 위같은 경우는 2차원배열

````python
a = tensor(3.,2); a
````

- tensor의 type()함수를 쓰게되면 float를 기본으로 하고있어 int로 입력해도 float형식으로나온다

````python
y = x@a + torch.rand(n)
````

````python
plt.scatter(x[:,0], y);
````

- 산점도(상관도)는  개의 연속형 변수 간의 관계를 보기 위하여 직교좌표의 평면에 관측점을 찍어 만든 통계 그래프이다.

````python
def mse(y_hat, y): return ((y_hat-y)**2).mean()
````

````python
a = tensor(-2.,4)
````

````python
y_hat = x@a
mse(y_hat, y)
````

````python
plt.scatter(x[:,0],y)
plt.scatter(x[:,0],y_hat);
````

## 9. 기울기 decent


````python
a = nn.Parameter(a); a
````

````python
def update():
    y_hat = x@a
    loss = mse(y, y_hat)
    if t % 10 == 0: print(loss)
    loss.backward()
    with torch.no_grad():
        a.sub_(lr * a.grad)
        a.grad.zero_()
````

````python
lr = 1e-1
for t in range(100): update()
````

````python

plt.scatter(x[:,0],y).detach().numpy()
plt.scatter(x[:,0],x@a).detach().numpy();
````
